```python
df["Masse_(10^24kg)"] = df["pl_bmasse"].apply(lambda x: x * 5.972)
df["Masse_(10^24kg)"]
```

```python
df["Gravity_(G_earth)"] = df["pl_bmasse"] * df["pl_rade"]
df["Gravity_(G_earth)"]
```

```python
df["Score_roch"] = df["pl_dens"] * (1/df["pl_rade"])**3
df["Score_roch"]
```

```python
df["pl_type"] = df["Score_roch"].apply(lambda x: "Rocheuse" if x > 3.5 else "Gazeuse/Mixte")
df["pl_type"].value_counts()
```

```python
df["sim_earth"] = (
    (1 - np.abs((df["pl_rade"] - 1) / (df["pl_rade"] + 1))) ** 0.57 *
    (1 - np.abs((df["pl_dens"] - 5.51) / (df["pl_dens"] + 5.51))) ** 1.07 *
    (1 - np.abs((df["Gravity_(G_earth)"] - 1) / (df["Gravity_(G_earth)"] + 1))) ** 0.70 *
    (1 - np.abs((df["pl_eqt"] - 288) / (df["pl_eqt"] + 288))) ** 5.58
) ** 0.25
```
