```python
def impute_planets(df, max_iter=10):
    df = df.copy()

    for _ in range(max_iter):
        before_na = df[["Gravity_G_earth", "pl_dens", "pl_rade"]].isna().sum().sum()

        # g = rho * R
        mask_g = df["Gravity_G_earth"].isna() & df["pl_dens"].notna() & df["pl_rade"].notna()
        df.loc[mask_g, "Gravity_G_earth"] = df.loc[mask_g, "pl_dens"] * df.loc[mask_g, "pl_rade"]

        # rho = g / R
        mask_d = df["pl_dens"].isna() & df["Gravity_G_earth"].notna() & df["pl_rade"].notna()
        df.loc[mask_d, "pl_dens"] = df.loc[mask_d, "Gravity_G_earth"] / df.loc[mask_d, "pl_rade"]

        # R = g / rho
        mask_r = df["pl_rade"].isna() & df["Gravity_G_earth"].notna() & df["pl_dens"].notna()
        df.loc[mask_r, "pl_rade"] = df.loc[mask_r, "Gravity_G_earth"] / df.loc[mask_r, "pl_dens"]

        after_na = df[["Gravity_G_earth", "pl_dens", "pl_rade"]].isna().sum().sum()

        # stop si plus aucun progrès
        if after_na == before_na:
            break

    return df
```

```python
def map_spectral_type(x):
    if pd.isna(x):
        return "Unknown"
    
    x = str(x).upper()

    if "WD" in x or "DQ" in x or "DC" in x:
        return "White Dwarf"

    if "O" in x:
        return "O"
    if "B" in x:
        return "B"
    if "A" in x:
        return "A"
    if "F" in x:
        return "F"
    if "G" in x:
        return "G"
    if "K" in x:
        return "K"
    if "M" in x:
        return "M"

    return "Other"


df["st_spectype_group"] = df["st_spectype"].apply(map_spectral_type)
```
