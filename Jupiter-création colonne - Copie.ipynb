```python
def impute_planets(df, max_iter=10):
    df = df.copy()

    for _ in range(max_iter):
        before_na = df[["Gravity_G_earth", "pl_dens", "pl_rade"]].isna().sum().sum()

        # g = rho * R
        mask_g = df["Gravity_G_earth"].isna() & df["pl_dens"].notna() & df["pl_rade"].notna()
        df.loc[mask_g, "Gravity_G_earth"] = df.loc[mask_g, "pl_dens"] * df.loc[mask_g, "pl_rade"]

        # rho = g / R
        mask_d = df["pl_dens"].isna() & df["Gravity_G_earth"].notna() & df["pl_rade"].notna()
        df.loc[mask_d, "pl_dens"] = df.loc[mask_d, "Gravity_G_earth"] / df.loc[mask_d, "pl_rade"]

        # R = g / rho
        mask_r = df["pl_rade"].isna() & df["Gravity_G_earth"].notna() & df["pl_dens"].notna()
        df.loc[mask_r, "pl_rade"] = df.loc[mask_r, "Gravity_G_earth"] / df.loc[mask_r, "pl_dens"]

        after_na = df[["Gravity_G_earth", "pl_dens", "pl_rade"]].isna().sum().sum()

        # stop si plus aucun progrès
        if after_na == before_na:
            break

    return df
```

```python
def map_spectral_type(x):
    if pd.isna(x):
        return "Unknown"
    
    x = str(x).upper()

    if "WD" in x or "DQ" in x or "DC" in x:
        return "White Dwarf"

    if "O" in x:
        return "O"
    if "B" in x:
        return "B"
    if "A" in x:
        return "A"
    if "F" in x:
        return "F"
    if "G" in x:
        return "G"
    if "K" in x:
        return "K"
    if "M" in x:
        return "M"

    return "Other"


df["st_spectype_group"] = df["st_spectype"].apply(map_spectral_type)
```

```python
def mass_limits(star):
    if star == "M":
        return 2.5, 4.0
    elif star == "K":
        return 1.0, 4.0
    elif star in ["G", "F"]:
        return 0.6, 4.0
    else:
        return None, None  # O, B, A → exclu


def compute_habitability(df):
    df = df.copy()

    # --------------------------
    # 1. Exclusion étoiles (O, B, A)
    # --------------------------
    df["star_allowed"] = ~df["st_spectype_group"].isin(["O", "B", "A"])

    # --------------------------
    # 2. Conditions de base
    # --------------------------
    rocky_ok = df["pl_type"] == "Rocheuse"
    water_ok = df["eau_possible"] == 1
    gravity_ok = df["Gravity_G_earth"] >= 0.8

    # --------------------------
    # 3. UTILISATION DIRECTE masse (corrigé)
    # --------------------------
    mass = df["pl_bmasse"]

    # --------------------------
    # 4. Limites dynamiques selon étoile
    # --------------------------
    min_mass, max_mass = zip(*df["st_spectype_group"].apply(mass_limits))
    df["min_mass"] = min_mass
    df["max_mass"] = max_mass

    df["mass_ok"] = (
        (mass >= df["min_mass"]) &
        (mass <= df["max_mass"])
    )

    # --------------------------
    # 5. Condition finale
    # --------------------------
    df["pt_habitable"] = (
        df["star_allowed"] &
        rocky_ok &
        water_ok &
        gravity_ok &
        df["mass_ok"]
    ).astype(int)

    return df
```
